```python
# Import necessary libraries
import cv2  # OpenCV for image and video handling
import mediapipe as mp  # Mediapipe for pose estimation

# Set up webcam capture
video_capture = cv2.VideoCapture(0)

# Initialize Mediapipe Pose detection and drawing utilities
mp_pose = mp.solutions.pose
pose_detector = mp_pose.Pose()
mp_draw = mp.solutions.drawing_utils

# Start reading video frames from the webcam
while True:
    success, frame = video_capture.read()

    if not success:
        print("Unable to read from webcam.")
        break

    # Convert the image from BGR (OpenCV default) to RGB (required by Mediapipe)
    frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Use Mediapipe to detect pose landmarks
    results = pose_detector.process(frame_rgb)

    # Print raw landmarks to console (useful for debugging)
    print(results.pose_landmarks)

    # If pose landmarks are found, draw them on the original image
    if results.pose_landmarks:
        mp_draw.draw_landmarks(frame, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)

    # Add a title text to the video feed
    cv2.putText(frame, "OnogwuEmma Pose Detection", (15, 55),
                cv2.FONT_HERSHEY_PLAIN, 2, (255, 255, 255), 2)

    # Show the frame in a window
    cv2.imshow("OnogwuEmma Pose Detection", frame)

    # Exit the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up: release the webcam and close the window
video_capture.release()
cv2.destroyAllWindows()
```
